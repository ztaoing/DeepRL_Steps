np.eye 是 NumPy 库中的一个函数，**用于创建单位矩阵。**

- 单位矩阵是一个方阵，其主对角线上的元素为 1，其余元素为 0。
- 单位矩阵在矩阵运算中非常重要，**因为它在矩阵乘法中起到类似于数字 1 的作用。**

```python
import numpy as np

# 创建一个 3x3 的单位矩阵
I = np.eye(3)

print(I)
```

```python
# 创建一个 3x4 的矩阵，主对角线为 1
I = np.eye(3, 4)

print(I)
```

```python
# 创建一个 3x3 的矩阵，对角线偏移 1
I = np.eye(3, k=1)

print(I)
```

```python
# 创建一个 3x3 的矩阵，对角线偏移 -1
I = np.eye(3, k=-1)

print(I)
```

```python
# 创建一个 3x3 的单位矩阵，数据类型为 int
I = np.eye(3, dtype=int)

print(I)
```
