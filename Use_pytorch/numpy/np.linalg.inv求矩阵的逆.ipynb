```python
import numpy as np

#  2x2 
A = np.array([[1, 2], [3, 4]])

# 
A_inv = np.linalg.inv(A)

print(" A:")
print(A)
print(" A :")
print(A_inv)
```

```python
#  A * A_inv = I
I = np.dot(A, A_inv)

print(" A  A :")
print(I)
print(np.eye(2))
```

```python
# 
# np.linalg.inv 
C = np.array([[1, 2], [2, 4]])
C_inv = np.linalg.inv(C)  # 
```

```python
# 
#  np.allclose 
np.allclose(np.dot(A, A_inv), np.eye(2))  #  True  False
```
