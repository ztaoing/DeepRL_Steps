```python
import torch
import torch.nn as nn

# 定义目标值和预测值
target = torch.tensor([1.0, 2.0, 3.0])
prediction = torch.tensor([1.5, 2.5, 3.5])

# 定义 MSE 损失函数
mse_loss = nn.MSELoss()

# 计算损失
loss = mse_loss(prediction, target)
print(f"MSE Loss: {loss.item()}")
```

```python
import torch
import torch.nn as nn
'''不同的 reduction 参数'''

# 定义目标值和预测值
target = torch.tensor([1.0, 2.0, 3.0])
prediction = torch.tensor([1.5, 2.5, 3.5])

# 定义 MSE 损失函数，使用不同的 reduction 参数
mse_loss_mean = nn.MSELoss(reduction='mean')
mse_loss_sum = nn.MSELoss(reduction='sum')
mse_loss_none = nn.MSELoss(reduction='none')

# 计算损失
loss_mean = mse_loss_mean(prediction, target)
loss_sum = mse_loss_sum(prediction, target)
loss_none = mse_loss_none(prediction, target)

print(f"MSE Loss (mean): {loss_mean.item()}")
print(f"MSE Loss (sum): {loss_sum.item()}")
print(f"MSE Loss (none): {loss_none}")
```
