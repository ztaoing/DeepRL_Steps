torch.uniform_() 是 PyTorch 中的一个函数，用于将张量的元素随机初始化为从均匀分布中采样的值。

这个函数可以直接在张量上进行原地操作（in-place），即直接修改张量的内容，而不是创建一个新的张量。

```python
from IPython.display import Image
```

```python
Image(filename='./Image/uniform.png',width=600)
```

```python
import torch

# 创建一个形状为 (2, 3) 的张量
tensor = torch.empty(2, 3)

# 使用均匀分布初始化张量，范围为 [0, 1)
tensor.uniform_(0, 1)
print(tensor)
```

```python
# 使用均匀分布初始化张量，范围为 [-1, 1)
tensor.uniform_(-1, 1)
print(tensor)
```

```python
torch.manual_seed(42)

tensor = torch.empty(2, 3)
tensor.uniform_(0, 1)
print(tensor)
```
