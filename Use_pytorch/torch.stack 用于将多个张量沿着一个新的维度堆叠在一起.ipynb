torch.stack 是 PyTorch 中的一个函数，用于将多个张量沿着一个新的维度堆叠在一起。它会创建一个新的维度，并将输入的张量沿着这个新维度进行堆叠。

**这在处理多个张量时非常有用，尤其是当你需要将它们组合成一个更高维度的张量时。**


torch.stack 非常适合以下场景：
- **组合多个张量**：将多个张量组合成一个更高维度的张量。
- **批量处理**：将多个样本的张量组合成一个批量张量。
- **多任务学习**：将多个任务的输出张量组合成一个张量，便于后续处理。

```python
# 示例 1：沿第 0 维堆叠
import torch

# 创建两个张量
tensor1 = torch.tensor([1, 2, 3])
tensor2 = torch.tensor([4, 5, 6])

print([tensor1, tensor2])
# 沿第 0 维堆叠
stacked_tensor = torch.stack([tensor1, tensor2], dim=0)
print(stacked_tensor)
```

```python
# 沿第 1 维堆叠
stacked_tensor = torch.stack([tensor1, tensor2], dim=1)
print(stacked_tensor)
```

```python
# 创建两个二维张量
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[5, 6], [7, 8]])

# 沿第 0 维堆叠
stacked_tensor = torch.stack([tensor1, tensor2], dim=0)
print(stacked_tensor)
```

```python
# 创建三个张量
tensor1 = torch.tensor([1, 2])
tensor2 = torch.tensor([3, 4])
tensor3 = torch.tensor([5, 6])

# 沿第 0 维堆叠
stacked_tensor = torch.stack([tensor1, tensor2, tensor3], dim=0)
print(stacked_tensor)
```

### np.vstack 只能沿着第一个维度堆叠

```python
import numpy as np

# 创建两个一维数组
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 垂直堆叠
result = np.vstack((a, b))

print(result)
```

```python
import numpy as np

# 创建两个二维数组
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# 垂直堆叠
result = np.vstack((a, b))

print(result)
```
